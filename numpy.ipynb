```python
# INDEXING, SLICING, ITERASI
a = np.arange(10)**2
print(a)
```

```python
# INVERS MATRIKS
a = np.array([[2,3], [1,1]])

# Invers dari a
a_inv = np.linalg.inv(a)

print(a)
print(a_inv)
```

```python
i = np.dot(a, a_inv)
print(i)
'''
TUGAS 4
JELASKAN KENAPA i MERUPAKAN MATRIKS IDENTITAS
'''
```
