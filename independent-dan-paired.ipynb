```r
# Soal nomer 1
# Uji normalitas
pbl <- c(40,60,60,90,70,75,76,80,85,50,75,68,70)
stad <- c(82,76,56,70,75,85,65,55,60,70,25)
shapiro.test(pbl)
shapiro.test(stad)
```

```r
kls_pbl <- rep(1, times=length(pbl))
kls_stad <- rep(2, times=length(stad))

nilai <- c(pbl, stad)
kelas <- c(kls_pbl, kls_stad)

bartlett.test(nilai, kelas)
```

```r
t.test(
    x = pbl,
    y = stad,
    alternative = 'g',
    paired = FALSE, # Ubah menjadi TRUE saat melakukan paired sample t-test
    var.equal = TRUE, # Ubah menjadi FALSE saat data tidak homogen
    conf.level = 0.95
)
```

```r
alpha = 0.05
dk = 22
titik_kritis <- qt(1-alpha, dk)
titik_kritis
```

t < titik kritis, sehingga H0 diterima
